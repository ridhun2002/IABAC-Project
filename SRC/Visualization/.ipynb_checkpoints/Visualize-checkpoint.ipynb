# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# LOADING DATA

```python
data = pd.read_excel("C:\\Users\\ashiq\\Desktop\\IABAC Project\\Data\\Raw\\INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
```

```python
data
```

```python
plt.figure(figsize=(25,25))
p=1
for i in data.iloc[:,:27]:
    ax=plt.subplot(7,4,p)
    sns.histplot(x=data[i],kde=True,hue=data.PerformanceRating,palette=['Green','Yellow'])
    p+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(25, 25), facecolor="white")
p = 1
for i in data.iloc[:,:27]:
        ax = plt.subplot(7, 4, p)
        sns.countplot(x=data[i], hue=data["PerformanceRating"])
        plt.xlabel(i, fontsize=25)
        plt.ylabel("PerformanceRating", fontsize=25)
        p+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EducationBackground",y="PerformanceRating",data=data)
plt.title("Education VS Performance")
plt.tight_layout
```

##### INSIGHTS
- BESIDES FROM MARKETING , LIFE SCIENCE , HUMAN RESOURCES , MEDICAL AND TECHNICAL DEGREES , OTHER EDUCATION BACKGROUND PEOPLES SHOWS MORE PERFORMANCE

```python
plt.figure(figsize=(10,5))
sns.barplot(x="MaritalStatus",y="PerformanceRating",data=data)
plt.title("Marital VS Performance")
plt.tight_layout
```

##### INSIGHT
- EVERY TYPES OF MARITALS SHOWING AROUND THE SAME PERFORMANCE

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EmpDepartment",y="PerformanceRating",data=data)
plt.title("Department VS Performance")
plt.xticks(rotation=45)
plt.tight_layout
```

##### INSIGHT
- DEVELOPMENT AND DATA SCIENCE DEPARTMENT SHOWING HIGH PERFORMANCE

```python
plt.figure(figsize=(10,5))
sns.barplot(x="EmpJobRole",y="PerformanceRating",data=data)
plt.title("Job Role VS Performance")
plt.xticks(rotation=45)
plt.tight_layout
```

##### INSIGHT
- ALL ARE SHOWING A NICE AMOUNT OF PERFORMANCE WITH SOME VARIATIONS

```python
plt.figure(figsize=(10,5))
sns.barplot(x="ExperienceYearsAtThisCompany",y="PerformanceRating",data=data)
plt.title("Experince VS Performance")
plt.tight_layout
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x="EmpHourlyRate",y="PerformanceRating",data=data)
plt.title("Hour Rate VS Performance")
plt.xticks(rotation=90)
plt.tight_layout
```
